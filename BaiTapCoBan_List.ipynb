```python
def find_min_and_index(lst):
    if not lst:
        return None, None
    min_val = lst[0]
    min_index = 0
    
    for i, val in enumerate(lst):
        if val < min_val:
            min_val = val
            min_index = i
            
    return min_val, min_index

# Thử
nums = [47, 15, 3, 38, 3, 50]
val, idx = find_min_and_index(nums)
print(f"Giá trị nhỏ nhất: {val}, Vị trí: {idx}")
# Kết quả: Giá trị nhỏ nhất: 3, Vị trí: 2
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
def is_palindrome(lst):
    return lst == lst[::-1]


print(is_palindrome([1, 2, 3, 2, 1])) # True
print(is_palindrome([1, 2, 3, 4, 5])) # False
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def move_zeros_to_front(lst):
  
    zeros = [x for x in lst if x == 0]
   
    others = [x for x in lst if x != 0]
    
  
    return zeros + others


nums = [1, 0, 3, 0, 12, 0, 5]
result = move_zeros_to_front(nums)
print(result) 
# Kết quả: [0, 0, 0, 1, 3, 12, 5]
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def check_sequence(lst):
    if len(lst) < 2:
        return "Tang"

    is_increasing = True
    is_decreasing = True

    for i in range(len(lst) - 1):
        if lst[i] < lst[i+1]:
            is_decreasing = False
        elif lst[i] > lst[i+1]:
            is_increasing = False 
        else:
            is_increasing = False
            is_decreasing = False
            break

    if is_increasing:
        return "Tang"
    elif is_decreasing:
        return "Giam"
    else:
        return "Vo danh"

# Kiểm tra thử
print(check_sequence([1, 3, 5, 6, 7, 8, 9]))  # Tang
print(check_sequence([9, 8, 7, 6, 5, 4, 0]))  # Giam
print(check_sequence([1, 2, 3, 7, 1, 0, 2]))  # Vo danh
print(check_sequence([1, 2, 2, 4, 5, 7]))     # Vo danh
```

### Bài 10: 
Nhập vào tên một người, in ra kí tự đầu tiên và kí tự cuối cùng của tên đó.
VD: "Nguyen Ai Quoc" => ["N", "c"]

```python
name = input("Nhập vào tên của bạn: ")


if name:
    first_char = name[0]
    last_char = name[-1]
    
    result = [first_char, last_char]
    print(f'"{name}" => {result}')
else:
    print("Bạn chưa nhập tên!")
```
