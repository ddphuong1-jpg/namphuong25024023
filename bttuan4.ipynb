W4A1: [SumOfNumbers]Đề bài: Tính tổng các số nguyên từ 1 đến n (n <= 1000).

```python
n = int(input("Nhập n: "))
tong = 0
for i in range(1, n + 1):
    tong += i
print(f"Tổng từ 1 đến {n} là: {tong}")
```

### Nhập vào một số tới khi được số nguyên dương thì thôi. Kiểm tra số đó có phải số nguyên tố không.

```python
# Nhập cho đến khi được số nguyên dương
while True:
    n = int(input("Nhập một số nguyên dương: "))
    if n > 0:
        break

# Kiểm tra số nguyên tố
is_prime = True
if n < 2:
    is_prime = False
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không phải là số nguyên tố")
```

Tính giai thừa của $n$ (0 < n < 100)

```python
n = int(input("Nhập n (0 < n < 100): "))
giai_thua = 1
for i in range(1, n + 1):
    giai_thua *= i
print(f"{n}! = {giai_thua}")
```

Đếm số chữ số của n (không dùng kiểu dữ liệu string).

```python
n = int(input("Nhập số nguyên n: "))
n = abs(n)
count = 0

if n == 0:
    count = 1
else:
    temp = n
    while temp > 0:
        temp //= 10 
        count += 1

print(f"Số chữ số là: {count}")
```

Tính tổng các số nguyên tố trong đoạn [a, b].

```python
def snt(num):
    if num < 2: return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0: return False
    return True

a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
tong_snt = 0

for i in range(a, b + 1):
    if snt(i):
        tong_snt += i

print(f"Tổng các số nguyên tố trong đoạn [{a}, {b}] là: {tong_snt}")
```
