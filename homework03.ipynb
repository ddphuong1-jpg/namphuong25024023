### Bài 4: In tam giác cân rỗng độ cao h

```python
h = int(input("Nhập h: "))
for i in range(1, h + 1):
    for j in range(1, 2 * h):
        if j == h - i + 1 or j == h + i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

### Bài 5: Tính tổng S = 1 + 2 + 3 + ... + n

```python
n = int(input("Nhập n: "))
s = 0
for i in range(1, n + 1):
    s += i
print("Tổng S =", s)
```

### Bài 6 & 7: In ước và đếm số lượng ước

```python
a = int(input("Nhập a: "))
dem = 0
print("Các ước số:", end=" ")
for i in range(1, a + 1):
    if a % i == 0:
        print(i, end=" ")
        dem += 1
print(f"\nSố lượng ước: {dem}")
```

### Bài 8: Ước chung của a và b

```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
print(f"Ước chung của {a} và {b}:", end=" ")
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

#### Bài 9: Kiểm tra số chính phương

```python
a = int(input("Nhập a: "))
is_cp = False
for i in range(a + 1):
    if i * i == a:
        is_cp = True
        break
print(f"{a} là số chính phương" if is_cp else f"{a} không phải số chính phương")
```

### Bài 10: Kiểm tra số hoàn hảo

```python
a = int(input("Nhập a: "))
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i
print(f"{a} là số hoàn hảo" if tong_uoc == a else f"{a} không phải số hoàn hảo")
```

### Bài 12: Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print(f"Số k tìm được là: {k} (với tổng S = {s})")
```

#### Bài 13: Tìm n nhỏ nhất để 1 + 1/2 + ... + 1/n > A

```python
A = float(input("Nhập A: "))
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print(f"Số n nhỏ nhất là: {n}")
```

####  Bài 14: Tìm Max, Min trong dãy số (dừng khi nhập -1)

```python
max_val = None
min_val = None

while True:
    num = int(input("Nhập số nguyên (nhập -1 để dừng): "))
    if num == -1:
        break
    
    if max_val is None or num > max_val:
        max_val = num
    if min_val is None or num < min_val:
        min_val = num

if max_val is not None:
    print(f"Số lớn nhất: {max_val}")
    print(f"Số nhỏ nhất: {min_val}")
else:
    print("Không có số nào được nhập.")
```

#### Bài 16 & 17: Đếm chữ số chẵn/lẻ và Tính tổng các chữ số

```python
n = int(input("Nhập n: "))
chan = 0
le = 0
tong = 0
temp = n

while temp > 0:
    chu_so = temp % 10 # Lấy chữ số cuối cùng
    tong += chu_so     # Cộng vào tổng (Bài 17)
    
    if chu_so % 2 == 0: # Kiểm tra chẵn lẻ (Bài 16)
        chan += 1
    else:
        le += 1
    temp //= 10        # Bỏ chữ số cuối cùng

print(f"Bài 16: {chan} chữ số chẵn, {le} chữ số lẻ")
print(f"Bài 17: Tổng các chữ số là {tong}")
```

#### Bài 18: Kiểm tra n có phải dạng 3^k hay không?

```python
n = int(input("Nhập n: "))
is_3k = False
if n > 0:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        is_3k = True

print(f"{n} {'là' if is_3k else 'không phải'} dạng 3^k")
```

#### Bài 19: Số Fibonacci lớn nhất không vượt quá A

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1
f_max = 1

while f2 <= A:
    f_max = f2
    f1, f2 = f2, f1 + f2 # Tính số tiếp theo trong dãy

print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f_max}")
```

#### Bài 20: Đếm các chữ số là số nguyên tố

```python
n = int(input("Nhập n: "))
dem_snt = 0
temp = n

while temp > 0:
    chu_so = temp % 10
    # Kiểm tra trực tiếp vì chữ số chỉ từ 0-9
    if chu_so in [2, 3, 5, 7]:
        dem_snt += 1
    temp //= 10

print(f"Số chữ số là số nguyên tố trong {n} là: {dem_snt}")
```
