```python
# W5A3: Kiểm tra số nguyên tố

def prime(n):          
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

while True:
    n = int(input())
    if n <= 0:
        print("False")
    else:
        if prime(n):
            print("True")
        else:
            print("False")
        break
                   
```

```python
#W5A4: Kiểm tra số hoàn hảo (Tổng ước số bằng chính nó)
def is_perfect_number(n: int):
    if n <= 0: 
        return False
    else:
          sum_divisors = sum(i for i in range(1, n) if n % i == 0)
          if sum_divisors == n:
              return True
   

n = int(input())
if is_perfect_number(n) == True:
    print("True")
else:
    print("False")
                   
```

Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
# W5A8: Khoảng cách Hamming (Số vị trí bit khác nhau)
def hamming_distance(x: int, y: int) -> int:
    # XOR x và y, sau đó đếm số bit 1 trong kết quả
    return bin(x ^ y).count('1')
```

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài.

```python
# W5A10: Kiểm tra hai từ đẳng cấu (Isomorphic)
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b): return False
    mapping = {}
    used_values = set()
    for char_a, char_b in zip(a, b):
        if char_a in mapping:
            if mapping[char_a] != char_b:
                return False
        else:
            if char_b in used_values:
                return False
            mapping[char_a] = char_b
            used_values.add(char_b)
    return True

a = input()
b = input()
print(is_isomorphic(a,b))
```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 


```python
# W5A12: Chia có dư sao cho 0 <= |r| < |divisor|
def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    if divisor == 0: return (0, 0)
    q = dividend // divisor
    r = dividend % divisor
    return q, r
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
# W5A15: Kiểm tra tam giác vuông (Định lý Pythagoras)
def is_right_triangle(a: float, b: float, c: float) -> bool:
    sides = sorted([a, b, c])
    # Kiểm tra điều kiện tam giác và định lý Pythagoras
    if sides[0] + sides[1] <= sides[2]: return False
    return math.isclose(sides[0]**2 + sides[1]**2, sides[2]**2)
```

```python
# W5A16: Kiểm tra số Armstrong
# (Số có tổng các lũy thừa bậc n của các chữ số bằng chính nó, với n là số chữ số)
# ==========================================
def is_armstrong(x: int) -> bool:
    if x < 0: return False
    s = str(x)
    n = len(s)
    return x == sum(int(digit)**n for digit in s)
```

```python
# W5A17: Kiểm tra số chính phương
# ==========================================
def is_perfect_square(x: int) -> bool:
    if x < 0: return False
    root = int(math.isqrt(x))
    return root * root == x
```

```python
# W5A18: Kiểm tra số đối xứng (chỉ dùng toán số học)
# ==========================================
def is_palindrome(x: int) -> bool:
    if x < 0: return False
    original = x
    reversed_num = 0
    while x > 0:
        reversed_num = reversed_num * 10 + x % 10
        x //= 10
    return original == reversed_num
```

```python
# W5A19: Kiểm tra số kỳ diệu 
# (Định nghĩa phổ biến: Tổng các chữ số lặp lại nhiều lần cuối cùng bằng 1)
# ==========================================
def is_magic_number(x: int) -> bool:
    if x <= 0: return False
    while x > 9:
        s = 0
        while x > 0:
            s += x % 10
            x //= 10
        x = s
    return x == 1
```

```python
# W5A23: Đếm số nghiệm của phương trình bậc 2
# ==========================================
def quadratic_roots_count(a, b, c) -> int:
    if a == 0:
        if b != 0: return 1
        return 0 if c != 0 else float('inf') # Vô số nghiệm nếu a=b=c=0
    
    delta = b**2 - 4*a*c
    if delta > 0: return 2
    elif delta == 0: return 1
    else: return 0
```

```python
# W5A24: Kiểm tra 3 điểm thẳng hàng
# ==========================================
def are_collinear(x1, y1, x2, y2, x3, y3) -> bool:
    # So sánh độ dốc bằng phép nhân chéo để tránh chia cho 0
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
```

```python
# W5A25: Kiểm tra điểm p nằm trong hoặc trên biên tứ giác
# ==========================================
def point_in_quadrilateral(quad: list, p: tuple) -> bool:
    # Sử dụng thuật toán Ray-Casting
    x, y = p
    n = len(quad)
    inside = False
    p1x, p1y = quad[0]
    for i in range(1, n + 1):
        p2x, p2y = quad[i % n]
        
        # Kiểm tra điểm có nằm trên đoạn thẳng biên không
        if min(p1x, p2x) <= x <= max(p1x, p2x) and min(p1y, p2y) <= y <= max(p1y, p2y):
            if (y - p1y) * (p2x - p1x) == (x - p1x) * (p2y - p1y):
                return True 
        
        # Ray casting cho điểm nằm trong
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xints = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xints:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside
```

```python

# W5A27: Số lượng chữ số khác 0 trong list số nguyên
# ==========================================
def count_nonzero_digits(nums: list) -> int:
    count = 0
    for num in nums:
        for digit in str(abs(num)):
            if digit != '0':
                count += 1
    return count
```

```python
# W5A28: Tính góc ABC (độ)
# ==========================================
def calculate_angle_abc(x1, y1, x2, y2, x3, y3) -> float:
    # Vector BA và BC
    BA = (x1 - x2, y1 - y2)
    BC = (x3 - x2, y3 - y2)
    
    dot_product = BA[0] * BC[0] + BA[1] * BC[1]
    mag_BA = math.sqrt(BA[0]**2 + BA[1]**2)
    mag_BC = math.sqrt(BC[0]**2 + BC[1]**2)
    
    if mag_BA == 0 or mag_BC == 0:
        return 0.0 # Bị trùng điểm
        
    cos_theta = dot_product / (mag_BA * mag_BC)
    cos_theta = max(min(cos_theta, 1.0), -1.0) # Tránh lỗi sai số dấu phẩy động
    return math.degrees(math.acos(cos_theta))
```

```python
# W5A33: Tính tiền taxi thông minh
# (Sử dụng các hằng số giả định vì đề không cho bảng giá cụ thể)
# ==========================================
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # GIÁ GIẢ ĐỊNH - Vui lòng thay đổi theo đề bài gốc của bạn
    OPENING_FEE = 10000        # Giá mở cửa
    KM_PRICE = 15000           # Giá mỗi km
    WAIT_FEE_PER_MIN = 1000    # Phí chờ mỗi phút
    
    total = OPENING_FEE + (km * KM_PRICE) + (wait_minutes * WAIT_FEE_PER_MIN)
    
    # Giả định phụ phí ban đêm (từ 22h đến 5h sáng): tăng 20%
    if 22 <= start_hour or start_hour <= 5:
        total *= 1.2
        
    # Giả định giảm giá: Đi trên 50km được giảm 10%
    if km > 50:
        total *= 0.9
        
    return int(total)
```

```python
# W5A34: Tổng các số có các chữ số khác nhau từ 1 đến n
# ==========================================
def has_unique_digits(x: int) -> bool:
    s = str(abs(x))
    return len(s) == len(set(s))

def sum_1_to_n(n: int) -> int:
    if n <= 0:
        return 0
    total = 0
    for i in range(1, n + 1):
        if has_unique_digits(i):
            total += i
    return total
```
