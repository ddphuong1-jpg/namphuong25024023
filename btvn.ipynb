### Problem 2
 Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
 diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
 chữ số phần thập phân. 


```python

a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))
pi = 3.14
s_hcn = a * b
r = a / 2
s_tron = pi * (r ** 2)
s_con_lai = s_hcn - s_tron
print(f"Diện tích còn lại để trồng cây là: {round(s_con_lai, 2)}")
```

### Problem 3
Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem
trước cấu trúc lựa chọn: if….else, ma) 

```python
# Bước 1: Nhập một kí tự từ bàn phím
c = input("Nhập vào một kí tự: ")

# Bước 2: Kiểm tra điều kiện bằng if...else
if c.isupper():
    # Nếu là chữ hoa, chuyển thành chữ thường
    ket_qua = c.lower()
    print(f"Kí tự sau khi chuyển đổi: {ket_qua}")
elif c.islower():
    # Nếu là chữ thường, chuyển thành chữ hoa
    ket_qua = c.upper()
    print(f"Kí tự sau khi chuyển đổi: {ket_qua}")
else:
    # Nếu không phải chữ cái (ví dụ: số, ký hiệu)
    print("Đây không phải là chữ cái trong bảng chữ cái.")
```

### Problem 6
 Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”

```python
import math
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if (a + b > c) and (a + c > b) and (b + c > a):
    p = (a + b + c) / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Diện tích của tam giác là {dien_tich:.2f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

### Problem 8
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng 


```python
def tien_dien(chi_so_thang_truoc, chi_so_thang_nay):
    so_dien = chi_so_thang_nay - chi_so_thang_truoc
    tien_truoc_thue = 0
    
    
    if so_dien <= 50:
        tien_truoc_thue = so_dien * 1984
    elif so_dien <= 100:
        tien_truoc_thue = 50 * 1984 + (so_dien - 50) * 2050
    elif so_dien <= 200:
        tien_truoc_thue = 50 * 1984 + 50 * 2050 + (so_dien - 100) * 2380
    elif so_dien <= 300:
        tien_truoc_thue = 50 * 1984 + 50 * 2050 + 100 * 2380 + (so_dien - 200) * 2998
    elif so_dien <= 400:
        tien_truoc_thue = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (so_dien - 300) * 3350
    else:
        tien_truoc_thue = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (so_dien - 400) * 3460
        
    
    tien_cuoi_cung = tien_truoc_thue * 1.08
    return int(tien_cuoi_cung)


ten = input("Nhập họ tên chủ hộ: ")
chi_so_thang_truoc = int(input("Chỉ số tháng trước: "))
chi_so_thang_nay = int(input("Chỉ số tháng này: "))


thanh_tien = tien_dien(chi_so_thang_truoc, chi_so_thang_nay)
print(f"Họ va tên: {ten}")
print(f"Tiền phải trả là: {thanh_tien} đồng")
```
