### Problem 5

```python
import math
a, b, c = map(float, input("Nhập a, b, c: ").split())
delta = b**2 - 4*a*c
if delta < 0: print("Vô nghiệm")
elif delta == 0: print(f"Nghiệm kép x = {-b/(2*a)}")
else:
    x1 = (-b + math.sqrt(delta))/(2*a)
    x2 = (-b - math.sqrt(delta))/(2*a)
    print(f"x1 = {x1}, x2 = {x2}")
```

### Problem 10


```python
x0, y0 = map(float, input("Tâm I(x0, y0): ").split())
r = float(input("Bán kính R: "))
xa, ya = map(float, input("Điểm A(x, y): ").split())
khoang_cach = math.sqrt((xa-x0)**2 + (ya-y0)**2)
print("A thuộc đường tròn" if khoang_cach <= r else "A nằm ngoài đường tròn")
```
