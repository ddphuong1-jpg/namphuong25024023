```python
## W6A16: Nhập và in ma trận n \times m$
n, m = map(int, input("Nhập n, m: ").split())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

for row in matrix:
    for val in row:
        print(f"{val:4d}", end="")
    print()
```

```python
## W6A19 & W6A22: Tìm vị trí giá trị lớn nhất
def vi_tri_max(lst):
    if not lst: return None
    max_val = max(lst)
    return lst.index(max_val)

# Ví dụ sử dụng:
# print(vi_tri_max([1, 5, 3, 5, 2])) -> Trả về 1
```

```python
## W6A20: Tìm vị trí đầu tiên của giá trị k
def tim_vi_tri_k(lst, k):
    try:
        return lst.index(k)
    except ValueError:
        return -1
```

```python
## W6A21: Dictionary ánh xạ chuỗi -> danh sách vị trí
s = input("Nhập chuỗi: ").split()
d = {}
for i, word in enumerate(s):
    if word not in d:
        d[word] = []
    d[word].append(i)

# Chuyển list sang tuple theo yêu cầu đề bài
d_final = {k: tuple(v) for k, v in d.items()}
print(d_final)
```

```python
## W6A23: Xóa các số < x trong mảng
def xoa_nho_hon_x(lst, x):
    # Sử dụng list comprehension để tạo list mới chỉ chứa các số >= x
    return [i for i in lst if i >= x]
```

```python
## W6A24: Đẩy tất cả số 0 về đầu mảng
def day_so_0_ve_dau(lst):
    zeros = [i for i in lst if i == 0]
    others = [i for i in lst if i != 0]
    return zeros + others

# Test: [0, 2, 3, 2, 4, 0, 0, 0, 9] -> [0, 0, 0, 0, 2, 3, 2, 4, 9]
```

```python
## W6A25: Kiểm tra giá trị từ 0-9 còn thiếu
def tim_so_thieu(lst):
    ton_tai = set(lst)
    thieu = [i for i in range(10) if i not in ton_tai]
    return thieu
```

```python
## W6A26: Tổng các phần tử cực đại địa phương
def tong_cuc_dai(lst):
    tong = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
    return tong
```

```python
## W6A28: Kiểm tra dãy tăng, giảm hay vô danh
def kiem_tra_day(lst):
    if len(lst) < 2: return "Vô danh"
    
    tang = tat_ca_giam = True
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]: tang = False
        if lst[i] <= lst[i+1]: tat_ca_giam = False
            
    if tang: return "Tăng thực sự"
    if tat_ca_giam: return "Giảm thực sự"
    return "Vô danh"
```

```python
# W6A9: Gộp hai dictionary và cộng giá trị
import ast

dict1 = ast.literal_eval(input("Nhập dictionary 1 (vd: {'a': 1, 'b': 2}): "))
dict2 = ast.literal_eval(input("Nhập dictionary 2 (vd: {'b': 3, 'c': 4}): "))

merged_dict = dict1.copy()
for key, value in dict2.items():
    merged_dict[key] = merged_dict.get(key, 0) + value

# Sắp xếp dict theo key tăng dần và in ra
sorted_dict = dict(sorted(merged_dict.items()))
print(sorted_dict)
```

```python
# W6A10: Tìm các cặp số có tổng bằng k
nums = list(map(int, input("Nhập danh sách số nguyên: ").split()))
k = int(input("Nhập số nguyên dương k: "))

pairs = set()
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            # Lưu cặp dưới dạng tuple đã sắp xếp để loại bỏ trùng lặp cặp (vd: (2,3) và (3,2))
            pairs.add(tuple(sorted((nums[i], nums[j]))))

# Sắp xếp các cặp theo phần tử đầu tiên tăng dần
sorted_pairs = sorted(list(pairs))
print(sorted_pairs)
```

```python
# W6A12: Tìm phần tử xuất hiện nhiều nhất (ưu tiên số nhỏ nhất)
nums = list(map(int, input("Nhập list các số nguyên: ").split()))

counts = {}
for num in nums:
    counts[num] = counts.get(num, 0) + 1

# Tìm số lần xuất hiện cao nhất
max_count = max(counts.values())

# Lọc các số có số lần xuất hiện bằng max_count
most_frequent_nums = [num for num, count in counts.items() if count == max_count]

# In ra số nhỏ nhất trong các số xuất hiện nhiều nhất
print(min(most_frequent_nums))
```
