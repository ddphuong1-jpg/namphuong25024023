### Bai 1: W3A1: In các chữ số đảo ngược của số nguyên n

```python
n = input("Nhập vào số nguyên n: ")
# Sử dụng slicing [::-1] để đảo ngược chuỗi
dao_nguoc = n[::-1]
print("Các chữ số đảo ngược là:", dao_nguoc)
```

### Problem 3
#### W3A3: Kiểm tra số có phải là lũy thừa của 2 không (Dùng bitwise)

```python
n = int(input("Nhập số cần kiểm tra: "))

if n > 0 and (n & (n - 1)) == 0:
    print(f"{n} là lũy thừa của 2")
else:
    print(f"{n} không phải là lũy thừa của 2")
```

### Problem 4
#### W3A4: Chia số m cho n, kết quả làm tròn xuống

```python
m = float(input("Nhập số bị chia m: "))
n = float(input("Nhập số chia n: "))

if n != 0:
    # Toán tử // thực hiện chia và làm tròn xuống
    ket_qua = m // n
    print(f"Kết quả {m} / {n} làm tròn xuống là: {int(ket_qua)}")
else:
    print("Không thể chia cho 0")
```

### Problem 6
#### W3A5: Chia m cho n, kết quả làm tròn lên

```python
import math

m = float(input("Nhập số bị chia m: "))
n = float(input("Nhập số chia n: "))

if n != 0:
    # math.ceil sẽ làm tròn lên số nguyên gần nhất
    ket_qua = math.ceil(m / n)
    print("Kết quả làm tròn lên là:", ket_qua)
else:
    print("Không thể chia cho 0")
```

### W3A11: Phân loại tam giác [Phân loại tam giác]

```python
a = int(input("Nhập cạnh a: "))
b = int(input("Nhập cạnh b: "))
c = int(input("Nhập cạnh c: "))

# Kiểm tra điều kiện hình thành tam giác trước
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

#### W3A13: Tính tiền điện

```python
kwh = float(input("Nhập số kWh tiêu thụ: "))
if kwh <= 50:
    tien = kwh * 1500
elif kwh <= 100:
    tien = 50 * 1500 + (kwh - 50) * 2000
else:
    tien = 50 * 1500 + 50 * 2000 + (kwh - 100) * 3000
print(f"Tổng tiền điện: {tien:.0f}đ")
```

#### W3A14: Giải phương trình bậc nhất ax + b = 0 (Làm tròn 2 chữ số)

#### W3A16: Làm tròn không dùng hàm có sẵn

```python
x = float(input("Nhập số thực x: "))
# Làm tròn lên
ceil_x = int(x) + (1 if x > int(x) else 0)
# Làm tròn xuống
floor_x = int(x)
# Làm tròn tới số nguyên gần nhất
round_x = int(x + 0.5) if x >= 0 else int(x - 0.5)

print(f"{ceil_x} {floor_x} {round_x}")
```

#### W3A17: Kiểm tra cấp số nhân

```python
a, b, c, d = map(float, input("Nhập 4 số a, b, c, d: ").split())
if a != 0 and b % a == 0:
    q = b // a
    if b * q == c and c * q == d:
        print("True")
    else:
        print("False")
else:
    print("False")
```
