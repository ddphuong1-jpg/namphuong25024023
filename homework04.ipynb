### W4A20: Kiểm tra n có phải dạng $2^k$ hay không

```python
n = int(input("Nhập n: "))
temp = n
is_power = True
if n <= 0:
    is_power = False
else:
    while n > 1:
        if n % 2 != 0:
            is_power = False
            break
        n //= 2
print(f"{temp} có dạng 2^k" if is_power else f"{temp} không phải dạng 2^k")
```

### W4A21: Tính tổng các chữ số của n

### W4A22: Đếm chữ số chẵn và lẻ của n

```python
n = input("Nhập n: ")
chan = 0
le = 0
for char in n:
    if int(char) % 2 == 0:
        chan += 1
    else:
        le += 1
print(f"Chẵn: {chan}, Lẻ: {le}")
```

#### W4A23: Tìm k sao cho S(k) = 1 + 2 + ... + k lớn nhất nhưng < n

```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print(f"k lớn nhất là: {k} (S({k}) = {s})")
```

### W4A24: Tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + ... + 1/n > A

```python
a = float(input("Nhập A: "))
tong = 0
n = 0
while tong <= a:
    n += 1
    tong += 1/n
print(f"n nhỏ nhất là: {n}")
```

### W4A25: Nhập dãy số cho đến khi gặp -1, in Min và Max

```python
max_val = -float('inf')
min_val = float('inf')
while True:
    n = int(input("Nhập số (-1 để dừng): "))
    if n == -1: break
    if n > max_val: max_val = n
    if n < min_val: min_val = n
print(f"Lớn nhất: {max_val}, Nhỏ nhất: {min_val}")
```

### W4A26: Tìm số Fibonacci lớn nhất nhưng không vượt quá A

```python
a = int(input("Nhập A: "))
f1, f2 = 1, 1
while f2 <= a:
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất không quá A là:", f1)
```

#### W4A27: Đếm số từ trong chuỗi (Không dùng split, len)

```python
s = input("Nhập chuỗi: ")
count = 0
in_word = False
for char in s:
    if char != " ":
        if not in_word:
            count += 1
            in_word = True
    else:
        in_word = False
print("Số từ:", count)
```

#### W4A28: In từ đầu tiên trong chuỗi (Không dùng split)

```python
s = input("Nhập chuỗi: ").strip()
first_word = ""
for char in s:
    if char == " ":
        break
    first_word += char
print("Từ đầu tiên:", first_word)
```

#### W4A29: Đổi hoa thành thường và ngược lại

```python
s = input("Nhập chuỗi: ")
res = ""
for char in s:
    if 'a' <= char <= 'z':
        res += chr(ord(char) - 32)
    elif 'A' <= char <= 'Z':
        res += chr(ord(char) + 32)
    else:
        res += char
print("Kết quả:", res)
```

#### W4A30: Tính tổng 3 số cách nhau bởi dấu phẩy

```python
s = input("Nhập 3 số (vd: 3, 12, 15): ")
tong = 0
num_str = ""
for char in s + ",": # Thêm dấu phẩy cuối để xử lý số cuối cùng
    if char == ",":
        tong += int(num_str.strip())
        num_str = ""
    else:
        num_str += char
print("Tổng:", tong)
```

#### W4A31: Đếm số ký tự in hoa, in thường và chữ số

```python
s = input("Nhập chuỗi: ")
hoa, thuong, so = 0, 0, 0
for char in s:
    if 'A' <= char <= 'Z': hoa += 1
    elif 'a' <= char <= 'z': thuong += 1
    elif '0' <= char <= '9': so += 1
print(f"Hoa: {hoa}, Thường: {thuong}, Số: {so}")
```
